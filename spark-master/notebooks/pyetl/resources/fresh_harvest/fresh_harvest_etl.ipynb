```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float, DateTime, ForeignKey, func
from sqlalchemy.orm import sessionmaker
import matplotlib.pyplot as plt
import mlflow
import mooncake
```

# Utils

```python
def normalize_str_columns(df, columns=None, lower=True, copy=True, trunc=None):
    """For each column in ``columns``, the following operations are applied:
    
    - NFKD unicode normalzation
    - ascii encoding
    - utf-8 decoding
    
    Parameters
    ----------
    df : pd.DataFrame
    
    columns : list of str, default=None
        List of columns to normalize. If None, all 'object' dtype columns
        are used.
    
    lower : bool, default=True
    
    copy : bool, default=True
    
    trunc: int, default=None
        If not None, truncates str columns to the given int.
    """
    if copy:
        df = df.copy()
    if columns is None:
        columns = df.select_dtypes('object').columns.tolist()
        
    def str_normalization(x):
        x = x.str.normalize('NFKD')
        x = x.str.encode('ascii', errors='ignore')
        x = x.str.decode('utf-8')
        if lower:
            x = x.str.lower()
        if trunc is not None:
            x = x.str.slice(0, trunc)
        return x
    
    df[columns] = df[columns].apply(str_normalization)
    return df

def create_dim(fact_table, columns, rename=None):
    """Creates dimensional table from fact table.
    
    Parameters
    ----------
    fact_table : pd.DataFrame
    
    columns : list of str.
        Columns used for dimensional table.
        
    rename : dict, str -> str
        Dictionary for renaming dimensional columns.
    """
    grouper = fact_table.groupby(columns).grouper
    dim = grouper.result_index.to_frame(index=False)
    if rename:
        dim.rename(columns=rename, inplace=True)
    return dim

def encode_relations(df, column_to_table, con, column_to_field=None, id_suffix=False, copy=True):
    """Encodes relationship columns.
    
    Parameters
    ----------
    df : pd.DataFrame
    
    column_to_table : dict, str -> class
        Dict containing the relationships to be encoded.
        E.g., {'product_line': ProductLine} means that the 
        'product_line' column inside ``df`` will be encoded with
        the primary key from the table ``ProductLine`` via an inner join
        with the 'name' field, i.e., ProductLine.name.

        If a field different from 'name' needs to be used, it must be passed
        using the ``column_to_field`` param.
    """
    if copy:
        df = df.copy()
    if column_to_field is None:
        column_to_field = {}

    for column, table in column_to_table.items():
        table_df = pd.read_sql(
            sql=session.query(table).statement, 
            con=con
        )
        if column in column_to_field:
            table_field = column_to_field[column]
        else:
            table_field ='name'
            
        map_ = dict(zip(table_df[table_field], table_df.id))
        df[column] = df[column].map(map_)
    
    if id_suffix:
        rename = {column: column + '_id' for column in column_to_table}
        df.rename(columns=rename, inplace=True)
            
    return df
```

# Schema

```python
# In general the connection str is: dialect+driver://username:password@host:port/database
# Note: The PostgreSQL dialect uses psycopg2 as the default DBAPI
# Note: The host is the name of the database service in docker-compose (postgres)
engine = create_engine('postgresql://postgres:admin@postgres/postgres', echo=True)
Base = declarative_base(bind=engine)
Session = sessionmaker(bind=engine)
session = Session()


class ProductLine(Base):
    __tablename__ = 'product_line'
    id = Column(Integer, primary_key=True, autoincrement=True)
    name = Column(String(50))
    
class Product(Base):
    __tablename__ = 'product'
    id = Column(Integer, primary_key=True, autoincrement=True)
    product_line_id = Column(Integer, ForeignKey('product_line.id'))
    name = Column(String(50))
    
class City(Base):
    __tablename__ = 'city'
    id = Column(Integer, primary_key=True, autoincrement=True)
    name = Column(String(50))
    
class Client(Base):
    __tablename__ = 'client'
    id = Column(Integer, primary_key=True, autoincrement=True)
    city_id = Column(Integer, ForeignKey('city.id'))
    name = Column(String(50))
    
class WorkForce(Base):
    __tablename__ = 'work_force'
    id = Column(Integer, primary_key=True, autoincrement=True)
    name = Column(String(50))
    
class FactTable(Base):
    __tablename__ = 'fact_table'
    id = Column(Integer, primary_key=True, autoincrement=True)
    product_id = Column(Integer, ForeignKey('product.id'))
    client_id = Column(Integer, ForeignKey('client.id'))
    work_force_id = Column(Integer, ForeignKey('work_force.id'))
    date = Column(DateTime)
    money = Column(Float)
    quantity = Column(Float)
    
#Base.metadata.create_all()
```

# Insert data

```python
# Load fact_table from csv
fact_table = pd.read_csv('ventas_2021.csv')

# Keep only database columns and rename them
columns_rename = {
    'Cliente': 'client',
    'Fecha del pedido': 'date',
    'Vendedor': 'work_force',
    'Importe total del pedido': 'money',
    'Cantidad': 'quantity',
    'productos': 'product',
    'Linea de productos': 'product_line',
    'Ciudad (facturaci√≥n)': 'city'
}
fact_table = fact_table[list(columns_rename)].rename(columns=columns_rename)

# Some preprocessing:
# normalize str columns and truncate them to 50 char
str_columns = ['client', 'work_force', 'product', 'product_line', 'city']
fact_table = normalize_str_columns(fact_table, str_columns, trunc=50)

# fill product_line nans with first word from product column
fact_table.loc[
    fact_table['product_line'].isna(), 
    'product_line'
] = fact_table.loc[
    fact_table['product_line'].isna(),
    'product'
].str.split(' ').str.get(0)
```

```python
# Product line dim

product_line_dim = create_dim(
    fact_table, 
    columns=['product_line'], 
    rename={'product_line': 'name'}
)
product_line_dim.to_sql(
    name='product_line',
    con=engine,
    index_label='id',
    if_exists='append'
)
```

```python
# Product dim

product_dim = create_dim(
    fact_table,
    columns=['product', 'product_line'],
    rename={'product': 'name'}
)

product_dim = encode_relations(
    product_dim,  
    column_to_table={'product_line': ProductLine}, 
    con=engine,
    id_suffix=True
)

product_dim.to_sql(
    name='product',
    con=engine,
    index_label='id',
    if_exists='append'
)
```

```python
# City dim

city_dim = create_dim(
    fact_table, 
    columns=['city'], 
    rename={'city': 'name'}
)
city_dim.to_sql(
    name='city',
    con=engine,
    index_label='id',
    if_exists='append'
)
```

```python
# Client dim

client_dim = create_dim(
    fact_table,
    columns=['client', 'city'],
    rename={'client': 'name'}
)

client_dim = encode_relations(
    client_dim,  
    column_to_table={'city': City}, 
    con=engine,
    id_suffix=True
)

client_dim.to_sql(
    name='client',
    con=engine,
    index_label='id',
    if_exists='append'
)
```

```python
# Work force dim

work_force_dim = create_dim(
    fact_table, 
    columns=['work_force'], 
    rename={'work_force': 'name'}
)
work_force_dim.to_sql(
    name='work_force',
    con=engine,
    index_label='id',
    if_exists='append'
)
```

```python
# Fact table encoding

column_to_table = {
    'client': Client,
    'work_force': WorkForce,
    'product': Product
}
cols_to_keep = [col + '_id' for col in column_to_table] + ['date', 'quantity', 'money']
encoded_fact_table = encode_relations(
    fact_table,
    column_to_table=column_to_table,
    id_suffix=True,
    con=engine
)[cols_to_keep]
encoded_fact_table['date'] = pd.to_datetime(encoded_fact_table['date'])

encoded_fact_table.to_sql(
    name='fact_table',
    con=engine,
    index_label='id',
    if_exists='append'
)
```

# Query fact table

```python
def get_product_sales(con, product_ids, client_ids=None):
    query = session.query(
        FactTable.product_id,
        FactTable.date,
        func.sum(FactTable.money).label('money'),
        func.sum(FactTable.quantity).label('quantity'),
    ).filter(FactTable.product_id.in_(product_ids))
    
    if client_ids is not None:
        query = query.filter(FactTable.client_id.in_(client_ids))
        
    query = query.group_by(FactTable.product_id, FactTable.date)
    return pd.read_sql(
        sql=query.statement,
        con=con
    )
```

```python
product_ids = [75, 27, 28]
df_sales = get_product_sales(engine, product_ids)
```

```python
df_sales_resample = df_sales.set_index('date'
                                      ).groupby('product_id'
                                      ).resample('D'
                                      ).agg({'money': sum, 'quantity': sum}
                                      ).reset_index()

groupby = df_sales_resample.groupby('product_id')
fig, axs = plt.subplots(len(groupby), 1, figsize=(15, 15))
for i, g in enumerate(groupby.groups):
    ax = axs[i]
    group = groupby.get_group(g).copy()
    group['price'] = group['money'] / group['quantity']
    group.set_index('date')[['quantity', 'price']].plot(secondary_y='price', ax=ax)
    
```

# Fit model

```python
from mooncake.nn import SeqToSeq
from mooncake.model_selection import train_test_split
from mooncake.preprocessing import (
    GroupTransformer, 
    TimeIndex, 
    CyclicalDates, 
    DataframeColumnTransformer
)
from mooncake.helper import column_selector, common_callbacks
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from torch.optim.lr_scheduler import OneCycleLR
```

```python
# Define X

product_ids = [75, 27, 28]
freq = 'D'
date = 'date'
group_ids = ['product_id']
target = 'quantity'
X = get_product_sales(engine, product_ids)
X = X.set_index(date
               ).groupby(group_ids
               ).resample(freq
               ).agg({'money': sum, 'quantity': sum}
               ).reset_index()
X.head()  # inspect X
```

```python
# Suppose we want to forecast the selected products from '2021-08-01' to '2021-10-01' (two months)

# Forecast period
test_start = '2021-09-01'
test_end = '2021-10-01'

# Sequence lengths
prediction_length = len(pd.date_range(test_start, test_end, freq=freq))
max_encoder_length = 2 * prediction_length
min_encoder_length = prediction_length

# Transform date column to time index
X['time_idx'] = X[date].copy()
time_idx_transformer = GroupTransformer(
    transformers=[('time_idx', TimeIndex(), 'time_idx')],
    group_ids=group_ids
)
X = time_idx_transformer.fit_transform(X)

# Add price column
X['price'] = X['money'] / X['quantity']
X.loc[X['price'].isna(), 'price'] = 0
del X['money']

# Change group_ids dtype
for g in group_ids:
    X[g] = X[g].astype(str)

# Train-test split 
X_train, X_test = train_test_split(
    X=X,
    test_start=test_start,
    test_end=test_end,
    date_column=date,
    freq=freq,
    sequence_length=max_encoder_length
)

X_train.head()  # inspect X_train
```

```python
def get_columns_segmentation(X, target, group_ids, time_idx):
    """Segments columns in three types:

    - time_varying_known_reals
        These include all numerical columns except ``target``, ``group_ids``
        and ``time idx``.

    - time_varying_unknown_reals
        Only includes ``target``

    - static_categoricals
        These include all 'object' dtype columns

    Returns
    -------
    dict : str -> list
    """
    from mooncake.helper import column_selector
    time_varying_known_reals = column_selector(
        dtype_include=[int, float],
        pattern_exclude=[target, time_idx] + group_ids
    )(X)
    time_varying_unknown_reals = [target]
    static_categoricals = column_selector(dtype_include=['object'])(X)
    return {
        'time_varying_known_reals': time_varying_known_reals,
        'time_varying_unknown_reals': time_varying_unknown_reals,
        'static_categoricals': static_categoricals
    }

def get_history(estimator, name):
    """Obtains history from estimator

    A history is any collection values recorded during training under a
    given name. Histories can be error metrics, loss functions, flags, etc

    For any fitted bluemeth estimator, histories are stored inside the
    `history` attribute inside the skorch model, e.g., estimator.net_.history.

    Parameters
    ----------
    estimator : bluemeth estimator
        Fitted blue meth estimator

    name : str
        Name of the history

    Returns
    -------
    history : list
    """
    if hasattr(estimator, 'estimators'):
        # Collect history in all estimators
        histories = []
        for est in estimator.estimators:
            histories.append(est.net_.history[:, name])

        # Get max length
        max_length = max(len(x) for x in histories)

        # Pad each history with nans at the end
        padded_histories = []
        for hist in histories:
            pad_hist = np.pad(
                hist,
                (0, max_length - len(hist)),
                constant_values=np.nan
            )
            padded_histories.append(pad_hist)

        # Return mean of all histories
        avg_histories = np.nanmean(padded_histories, axis=0)
        return avg_histories

    history = estimator.net_.history[:, name]
    return history

def log_array(key, array):
    """Logs array to current mlflow run

    Parameters
    ----------
    key : str
        Metric name

    array : array_like
        Collection of values to be log
    """
    for i, value in enumerate(array):
        mlflow.log_metric(key=key, value=value, step=i)
```

```python
# Preprocessing

# inner
selector = column_selector(
    dtype_include=['float', 'int'],
    pattern_exclude=group_ids + ['time_idx']
)
group_transformers = [
    ('cont', MinMaxScaler(), selector)
]

# outer
outer_transformers = [
    ('cyclical_dates', CyclicalDates(dayofweek=True, month=False), date)
]

# preprocessor pipeline
steps = [
    ('inner', GroupTransformer(group_transformers, group_ids)),
    ('outer', DataframeColumnTransformer(outer_transformers))
]
preprocessor = Pipeline(steps=steps)
X_train_transformed = preprocessor.fit_transform(X_train)
```

```python
# Fit

# column segmentation
column_segmentation = get_columns_segmentation(
    X_train_transformed, 
    target=target, 
    group_ids=group_ids, 
    time_idx='time_idx'
)

# callbacks
lr_scheduler = {
    'policy': OneCycleLR,
    'step_every': 'batch',
    'max_lr': 1e-3,
    'steps_per_epoch': 'iterations',
    'epochs': 'max_epochs'
}
callbacks = common_callbacks(
    lr_scheduler, 
    gradient_clipping=True,
    early_stopping=True, 
    patience=50
)


# model instance
estimator = SeqToSeq(
    group_ids=group_ids,
    time_idx='time_idx',
    target=target,
    max_prediction_length=prediction_length,
    max_encoder_length=max_encoder_length,
    min_encoder_length=min_encoder_length,
    callbacks=callbacks,
    emb_dim=50,
    tf_ratio=0.1,
    batch_size=16,
    max_epochs=100,
    hidden_size=32,
    **column_segmentation
)
try:
    # Try to pass test data as validation
    X_test_transformed = preprocessor.transform(X_test)
    ds_test = estimator.get_dataset(X_test_transformed)
    estimator.cv_split = ds_test
except (AssertionError, ValueError):
    # Here, we assume the reason ``X_test`` couldn't be transformed
    # to a dataset was because the target column contains NaN values.
    X_test[target] = X_test[target].fillna(0)
```

```python
# training

estimator.fit(X_train_transformed)
```

```python
# Log to mflow

# create experiment
with mlflow.start_run():

    # log preprocessor and estimator in a pipeline format
    steps = [('preprocessor', preprocessor), ('estimator', estimator)]
    pipe = Pipeline(steps=steps)
    mlflow.sklearn.log_model(pipe, 'pipe')

    # log estimator params
    params = estimator.get_params()
    params = {k: v for k, v in params.items()
              if k not in column_segmentation and k not in
              ('callbacks', 'callbacks_')}
    mlflow.log_params(params)

    # log losses
    train_loss = get_history(estimator, 'train_loss')
    valid_loss = get_history(estimator, 'valid_loss')
    log_array('train_loss', train_loss)
    log_array('valid_loss', valid_loss)
```

```python
yhat_train = estimator.predict(
    X_train_transformed, 
    raw=False, 
    inverse_transformer=preprocessor
)
yhat_test = estimator.predict(
    X_test_transformed, 
    raw=False, 
    inverse_transformer=preprocessor
)

groupby_train = yhat_train.groupby(group_ids)
groupby_test = yhat_test.groupby(group_ids)
fig, axs = plt.subplots(len(groupby), 2, figsize=(22, 15))
axs = axs.flatten()
for i, g in enumerate(groupby.groups):
    g = str(g)
    ax_l = axs[2 * i]
    ax_r = axs[2 * i + 1]
    
    groupby_train.get_group(g).set_index('time_idx').plot(ax=ax_l)
    groupby_test.get_group(g).set_index('time_idx').plot(ax=ax_r)
    
```
